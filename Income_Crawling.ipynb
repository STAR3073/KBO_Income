```python
# 분석 모듈
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
from datetime import datetime

# 크롤링 모듈
from selenium import webdriver
from bs4 import BeautifulSoup
import re

import os
os.chdir('C:/BigData/SooDa/2021/2021_과제_3')
```

```python
# 저장해둔 데이터 불러오기
data = pd.read_csv("data_1982_2020.csv", index_col=0)
```

```python
# 2001시즌부터 2020년시즌까지의 선수들의 명단 추출
data_2001_2020 = data[data['시즌']>=2001]
namebirth = data_2001_2020.loc[:, ['이름', '생일']] # 동명이인의 경우 생일로 구분
namebirth = namebirth.drop_duplicates(['이름'], keep = 'first') # 중복 제거
count = len(namebirth['이름']) # 총원 수

a = namebirth['이름']
b = namebirth['생일']
name = list(a)
birth = list(b)
```

```python
season=  []
income = []
na = []
bi = []
annual = []

# 크롬 드라이버 실행
driver = webdriver.Chrome('C:/Temp/chromedriver_win32/chromedriver.exe')

for i in range(count):
    
    url = 'http://www.statiz.co.kr/player.php?opt=11&name={}&birth={}'.format(name[i], birth[i])

    driver.get(url)
    driver.implicitly_wait(7)
    
    html = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div/section[2]/div/div[2]/div/div/div/div[2]/table/tbody').get_attribute("innerHTML") # 기록 table을 str형태로 저장
    soup = BeautifulSoup(html, 'html.parser') # str 객체를 BeautifulSoup 객체로 변경

    temp = [i.text.strip() for i in soup.findAll("td")] # td 태그에서, text만 저장하기
    temp = pd.Series(temp) # list 객체에서 series 객체로 변경
    
    for j in range(len(temp)):
        
        if j%3==0:
            na.append(name[i])
            bi.append(birth[i])
            season.append(int(temp[j]))
            annual.append(int(temp[j]) - int(temp[0]) + 1) 
            
        elif j%3==1:
            temp[j] = temp[j].replace(",", "")
            income.append(temp[j]) 
            
    print(i, name[i], birth[i], "완료")

# webdriver 종료
driver.close()

print("크롤링 완료")

# 데이터 프레임 생성 및 저장
df_income = pd.DataFrame({'이름' : na, '생일' : bi, '시즌' : season, '연봉(만원)' : income, '연차' : annual})
df_income.to_csv('income.csv', encoding="utf-8-sig")

print("데이터 프레임 생성 및 저장 완료")

print("최종 완료")
```
